```python
import pandas as pd 
import matplotlib.pyplot as plt 

# Dummy ranking (based on RQ2 output manually or reuse) 

df_table = pd.DataFrame({ 
    "Model": ["Linear","Random Forest","SVM"], 
    "R2": [0.65,0.85,0.78] 
})
df_table["Rank"] = df_table["R2"].rank(ascending=False)
df_table
```

```python
df_table.to_csv('/kaggle/working/table_rq5.csv', index=False)
print("table saved")
```

```python
plt.figure() 
plt.plot(df_table["Model"], df_table["Rank"], marker='o') 
plt.title("Model Ranking by Metric") 
plt.savefig('/kaggle/working/figure_rq5.pdf') 
plt.show()
```
